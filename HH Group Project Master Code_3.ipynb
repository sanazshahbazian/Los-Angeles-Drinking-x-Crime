```python
# import dependencies
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
import os 
import pandas as pd 

sns.set()
```

```python
# name file
file = 'Crime Data/Intoxication Data_2019_v3.csv'
```

```python
# Import our data into pandas from CSV
df1= pd.read_csv(file)
df1.head()
```

```python
# check if there are any null values in data frame
data = df1.isnull().sum() / len(df1)
data
```

```python
# check if 'Time' has any null values in column
df1[['Time']].isnull().sum() / len(df1)
```

```python
#filter by DRUNK DRIVING ALCOHOL/DRUGS
drunks = df1[df1['Charge Description'] == 'DRUNK DRIVING ALCOHOL/DRUGS']
drunks.head()
```

```python
#calculate the percent of male perpetrators

male=len(df1[df1['Sex Code'] == 'M']) / len(df1)

male*100
```

```python
#calculate the percent of female perpetrators
female=len(df1[df1['Sex Code'] == 'F']) / len(df1)
female*100
```

# Slide 9 - Crime by Gender and Location

```python
#create figure to showcase crime by male vs. female in each area of LA
plt.figure(figsize=(30,20))

sns.countplot(x=drunks['Area Name'],order=drunks['Area Name'].value_counts().index, hue=drunks['Sex Code']).set_title("Crime by Gender and location", fontsize=70)

plt.xticks(rotation=90)
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)

plt.xlabel('Area Name',fontsize=60)
plt.ylabel('Count',fontsize=60)

plt.savefig("Figures/Crime by Gender and location.png",bbox_inches='tight')
```

# Slide 10 - Arrest Rate based on Cities in LA

```python
#create figure to showcase arrest rate for each city in LA
plt.figure(figsize=(30,20))

plt.yticks(fontsize=50)
sns.countplot(y=df1['Area Name'],order=df1['Area Name'].value_counts().index[:10], palette='RdGy').set_title("Arrest rate based on cities", fontsize=40)
plt.savefig("Figures/Crime Base on Area Name",bbox_inches='tight')
plt.xlabel('count',fontsize=50)
```
